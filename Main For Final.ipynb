```python
import random

hats = [
    "no hat",
    "black beanie",
    "backwards yankee hat",
    "headphones",
    "sunglasses"
]

tops = [
    "basketball jersey",
    "black t-shirt",
    "workout shirt",
    "red t-shirt",
    "blue hoodies",
    "white button-up"
]

bottoms = [
    "blue jeans",
    "black athletic shorts",
    "black jeans",
    "khaki shorts",
    "grey sweats"
]

shoes = [
    "cowboy boots",
    "white Nike Air Max 270s",
    "black Dunk Lows",
    "Birkenstock sandals",
    "slippers"
]

def pick_outfit():
    hat = random.choice(hats)
    top = random.choice(tops)
    bottom = random.choice(bottoms)
    shoe = random.choice(shoes)

    print("Today's outfit:")
    print("hat:", hat)
    print("Top:", top)
    print("Bottom:", bottom)
    print("Shoes:", shoe)

pick_outfit()
```
